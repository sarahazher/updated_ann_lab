```python
import numpy as np



# defibe mc chulloch pitts neuron
def mp_neuron(inputs, weights, threshold):
# Compute the dot product of the inputs and weights
    net = np.dot (inputs, weights)
# Apply the threshold function
    if net >= threshold:
        output = 1
    else:
        output = 0
    return output

# def andnot neural network
def andnot_nn(inputs) :
    weights = [-2, 1]
    threshold = 0
# Compute the outputs of the two neurons in the first layer
    outputs = [mp_neuron (inputs, weights, threshold) for weights in [[-1, 1],[1,-1]]]
# Compute the final output of the ANDNOT function
    if outputs [0]== 1 and outputs [1] == 0:
        output =1                                                            
    else:
        output = 0
    return output

# test the andnot neural network

inputs = np. array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs = []
for i in range(inputs.shape[0]):
    y = andnot_nn(inputs[i].reshape(1, -1))
    outputs.append(y)
print("Inputs:", inputs)
print ("Outputs:", outputs)
```

```python
Objective:
The objective of this practical is to implement the ANDNOT logic function using a McCulloch-Pitts neural network model in Python.
By constructing a neural network based on the principles of the McCulloch-Pitts neuron, the aim is to create a computational model 
that can correctly compute the ANDNOT operation, which produces an output of 1 only when the first input is 1 and the second input is
0, otherwise outputting 0.

Theory in Detail (Simple Words):
A McCulloch-Pitts neuron is a simplified model of a biological neuron. It takes inputs, multiplies them by certain weights, adds them up, 
and compares the result to a threshold. If the result exceeds the threshold, the neuron "fires" and produces an output of 1; otherwise,
it outputs 0. In this practical, we're using McCulloch-Pitts neurons to create a neural network that can perform the ANDNOT operation.

To implement the ANDNOT function, we need to construct a neural network with two layers. The first layer consists of two McCulloch-Pitts 
neurons, each with specific weights and a threshold. These neurons process the input values and pass their outputs to the next layer. The 
second layer computes the final output of the ANDNOT function based on the outputs of the first layer.

In the ANDNOT operation, the output is 1 only when the first input is 1 and the second input is 0; otherwise, the output is 0. By setting 
appropriate weights and thresholds for the neurons in the first layer, we can ensure that the neural network correctly performs this logic operation.

Applications:
The ANDNOT function is a fundamental logic operation with various applications in digital circuits, computer science, and artificial intelligence.

    It is commonly used in logical operations, pattern recognition, and decision-making tasks. Applications include circuit design, digital signal
processing, pattern classification, and boolean logic operations in computer programs.

Algorithm:

Define the McCulloch-Pitts neuron function (mp_neuron) that takes inputs, weights, and a threshold.
Calculate the net input to the neuron by taking the dot product of inputs and weights.
If the net input exceeds the threshold, output 1; otherwise, output 0.
Define the ANDNOT neural network function (andnot_nn) that takes inputs.
Set weights and threshold for the neurons in the first layer.
Compute the outputs of the neurons in the first layer based on the inputs, weights, and threshold.
Determine the final output of the ANDNOT function based on the outputs of the first layer.
Test the ANDNOT neural network with different input combinations to verify its correctness.


A McCulloch-Pitts neuron is like a tiny decision-maker in a neural network. It gets signals from other neurons or inputs, weighs those signals based on their importance, adds them up, and then decides whether to "fire" or not. Firing means it sends a signal, which we represent as 1, while not firing means no signal, represented as 0.

To create an ANDNOT function, we need a network of these neurons working together. Imagine the first layer of neurons like gatekeepers. Each gatekeeper checks a specific aspect of the input signals and decides whether it's important enough to let through. For ANDNOT, we have two gatekeepers, each with their own idea of what's important. One might prioritize when the first input is 1, while the other cares about when the second input is 0.

The inputs we give to the network are combinations of 0s and 1s. Each combination represents a different scenario, like "Is it sunny (1) and hot (1)?" or "Is it raining (0) and cold (0)?" The network's job is to tell us whether a certain scenario fits the ANDNOT criteria: first input is 1, second input is 0.

The network goes through each scenario, and for each, the gatekeepers decide if it matches the criteria. If it does, the final decision is 1; if not, it's 0. So, for the "Is it sunny and not hot?" scenario, the network would return 1, indicating it matches the ANDNOT condition.

Applications:

The ANDNOT function might sound simple, but it's incredibly useful in computing and decision-making. For example, in electronics, it's used to build circuits that perform specific tasks based on input conditions. In software, it helps filter and sort data, making decisions based on certain criteria. Additionally, in artificial intelligence, logical operations like ANDNOT are fundamental building blocks for creating more complex decision-making systems.

Algorithm:

McCulloch-Pitts Neuron:
Calculate the weighted sum of inputs, reflecting their importance.
Decide whether to fire based on a threshold. If the weighted sum is higher than the threshold, fire; otherwise, don't.
ANDNOT Neural Network:
Set up the first layer of neurons with specific criteria for firing.
Process input combinations through the first layer, letting the gatekeepers decide if each combination matches the ANDNOT condition.
Combine the decisions of the gatekeepers to give the final output of the ANDNOT function.
Testing:
Input different scenarios to the network and check if the output matches what's expected for the ANDNOT function. If it does, the network is working correctly.
```
